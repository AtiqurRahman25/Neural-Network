```python
import torch

x = torch.tensor([1.0, 2.0], dtype=torch.int32)
y = torch.tensor([3.0, 4.0], dtype=torch.int32)

z = x + y
print(z)
```

```python
x =  torch.tensor(2.0, requires_grad=True)
y = 9*x**4 + 2*x**3 + 3*x**2 + 6*x + 1
y.backward()
x.grad
print(x.grad)
```

```python
import torch.nn as nn

model = nn.Sequential(nn.Linear(10, 5),
                      nn.ReLU(),
                      nn.Linear(5, 1),
                      nn.Sigmoid())


# Forward pass
input = torch.randn(1, 10)  # Random input
output = model(input)
print(output)
```
